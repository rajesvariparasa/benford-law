```python
import pandas as pd
```

```python
filepath = "C:/Users/Rajesvari Parasa/Desktop/districts.csv"
df = pd.read_csv(filepath)
df.head(10)
```

```python
#creating dailies from cumus
df['daily_new'] = df.sort_values('Date').groupby(['District', 'State'])['Confirmed'].diff().fillna(df['Confirmed'])
df['daily_recovered']= df.sort_values('Date').groupby(['District', 'State'])['Recovered'].diff().fillna(df['Recovered'])
df['daily_deaths'] = df.sort_values('Date').groupby(['District', 'State'])['Deceased'].diff().fillna(df['Deceased'])
```

```python
df.sort_values(['State', 'District', 'Date']).iloc[2500:2550]
```

```python
#converting to string

df1 = df.astype({'daily_new': 'str', 'daily_recovered': 'str', 'daily_deaths': 'str' })
df1.dtypes
```

```python
def extractfirstdigit_as_str(t):
    stringnumber = t[0:1]
    return(stringnumber)
```

```python
df1['ld_cases'] = df1.apply(lambda x: extractfirstdigit_as_str(x['daily_new']),axis=1)
df1['ld_rec'] = df1.apply(lambda x: extractfirstdigit_as_str(x['daily_recovered']),axis=1)
df1['ld_dec'] = df1.apply(lambda x: extractfirstdigit_as_str(x['daily_deaths']),axis=1)
```

```python
df1
```

```python
#converting back to float
df1 = df1.astype({'daily_new': 'float64', 'daily_recovered': 'float64', 'daily_deaths': 'float64' })
df1.dtypes
```

```python
df1.ld_cases.unique()
```

```python
df2 = df1.ld_cases.value_counts()
df3 = df1.ld_rec.value_counts()
df4  = df1.ld_dec.value_counts()
```

```python
df5 = pd.concat([df2, df3, df4], axis=1)
```

```python
df5
#dataframe with counts of leading digits for daily counts of deaths, cases, recovered
```

# State wise analysis

```python
df6 = df1.groupby('State').ld_cases.value_counts()
df7 = df1.groupby('State').ld_rec.value_counts()
df8 = df1.groupby('State').ld_dec.value_counts()

df9 = pd.concat([df6,df7,df8],axis=1)
df9
```
